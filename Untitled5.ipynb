```python
def Bright_Eyes():
    Jingming_data = {
        '定位': ' 在面部，目內眥角稍上方凹陷處。',
        '解剖': '在眶內緣瞼內側韌帶中，深部為眼內直肌；有內眥動、靜脈和滑車上下動、靜脈，深層上方有眼動、靜脈本干；布有滑車上、下神經，深層為眼神經，上方為鼻睫神經。',
        '主治': '目赤腫痛，流淚，視物不明，目眩，近視，夜盲，色盲。',
        '配伍': '配球後、光明治視目不明。',
        '刺灸法': '囑患者閉目，醫者左手輕推眼球向外側固定，左手緩慢進針，緊靠眶緣直刺0.5～1寸。不捻轉，不提插（或只輕微地捻轉和提插）。出針後按壓針孔片刻，以防出血。本穴禁灸。'
    }
    return Jingming_data

def  Gathered_Bamboo():
    Zanzhu_data = {
        '定位': '在面部，當眉頭陷中，眶上切跡處。',
        '解剖': '有額肌及皺眉肌；當額動、靜脈處；布有額神經內側支。',
        '主治': '頭痛，口眼歪斜，目視不明，流淚，目赤腫痛，眼瞼（目閏）動，眉棱骨痛，眼瞼下垂。',
        '配伍': '配陽白治口眼歪斜、眼瞼下垂。',
        '刺灸法': '平刺0.5～0.8寸。禁灸。'
    }
    return Zanzhu_data


def  Eyebrows_Pouring():
    Meichong_data = {
        '定位': '在頭部，當攢竹直上入髮際0.5寸，神庭與曲差連線之間。',
        '解剖': '有額肌；當額動、靜脈處；布有額神經內側支。',
        '主治': '頭痛，眩暈，鼻塞，癲癇。',
        '配伍': '配太陽治頭痛。',
        '刺灸法': '平刺0.3～0.5寸。'
    }
    return Meichong_data

def  Crooked_Curve():
    Quchai_data = {
        '定位': '在頭部，當前髮際正中直上0.5寸，旁開1.5寸，即神庭與頭維連線的內1/3與中1/3交點。',
        '解剖': '有額肌；當額動、靜脈處；布有額神經內側支。',
        '主治': ' 頭痛，鼻塞，鼽衄，目視不明。',
        '配伍': ' 配合谷治頭痛、鼻塞。',
        '刺灸法': '平刺0.5～0.8寸。'
    }
    return Quchai_data

def Fifth_Place():
  Wuchu_data = {
      '定位':'在頭部，當前髮際正中直上1寸，旁開1.5寸。',
      '解剖':'有額肌;當額動、靜脈處;布有額神經內側支。',
      '主治':'頭痛，目眩、癲癇。',
      '配伍':'配合谷、太衝治頭痛、目眩。',
      '刺灸法':'平刺0.5~0.8寸。'
  }
  return Wuchu_data

def Bladder_Meridian_of_Foot_Taiyang():
    acupoints_info = {
        '睛明穴': Bright_Eyes(),
        '攢竹穴': Gathered_Bamboo(),
        '眉衝穴': Eyebrows_Pouring(),
        '曲差穴': Crooked_Curve(),
        '五處穴': Fifth_Place()
        # ... 其他穴位
    }
    return acupoints_info

def get_acupoint_info(acupoint_name, info_category=None):
    acupoints_info = Bladder_Meridian_of_Foot_Taiyang()

    acupoint_data = acupoints_info.get(acupoint_name, None)
    if acupoint_data is None:
        return "未找到指定的穴位信息。"

    if info_category:
        return acupoint_data.get(info_category, "未找到指定類別的信息。")

    return acupoint_data

def query_acupoint():
    while True:
        print("\n--- 查詢系統 ---")
        acupoint_name = input("請輸入您要查詢的穴位名稱（輸入'退出'結束查詢）:\n(睛明穴、攢竹穴、眉沖穴、曲差穴、五處穴選擇其一owo):")
        if acupoint_name == "退出":
            print("\n--- 已退出查詢 ---")
            print("感謝使用本系統，祝你好運owo")
            break

        acupoint_info = get_acupoint_info(acupoint_name)
        if "未找到" in acupoint_info:
            print(f"\n{acupoint_info}\n")
            continue

        while True:
            print(f"\n找到穴位：{acupoint_name}")
            print("可查詢的信息類別有：定位、解剖、主治、配伍、刺灸法")
            info_category = input("請輸入您要查詢的信息類別（輸入'全部'顯示所有信息，或輸入'返回'返回上一層）:")

            if info_category == "返回":
                print("\n返回上一層...\n")
                break
            elif info_category == "全部":
                print(f"\n{acupoint_name} - 全部信息:")
                for category, detail in acupoint_info.items():
                    print(f"{category}: {detail}")
                print()  # 為了排版整潔owo
            elif info_category in ["定位", "解剖", "主治", "配伍", "刺灸法"]:
                category_info = get_acupoint_info(acupoint_name, info_category)
                print(f"\n{acupoint_name} - {info_category}:\n{category_info}\n")
            else:
                print("\n輸入的信息類別不正確，請重新輸入。\n")

            # 再次查詢或返回
            while True:
                next_action = input("輸入'再查詢'以查詢此穴位的其他信息，或輸入'返回'返回穴位選擇：")
                if next_action in ["再查詢", "返回"]:
                    break
                else:
                    print("請輸入有效的指令（'再查詢' 或 '返回'）。")
            if next_action == "返回":
                print("\n返回穴位選擇...")
                break
# 啟動查詢
query_acupoint()
```
